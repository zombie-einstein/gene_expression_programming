```python
r = np.arange(0,1,0.01)
```

```python
from itertools import product
c = np.array(list(product(r, r))).T.astype('float32')
```

```python
f_map = {'A': {'f': np.add, 'n': 2},
         'M': {'f': np.multiply, 'n': 2},
         'D': {'f': np.subtract, 'n': 2},
         'C': {'f': np.cos, 'n': 1},
         'S': {'f': np.sin, 'n': 1}}
```

```python
foo = numba_genome.GEP(f_map, 2, 10, float32[:,:])
```

```python
g = [foo.random_genome() for _ in range(10)]
```

```python
bar = [foo.phenotype(i) for i in g]
```

```python
res = [b[0](c) for b in bar]
```

```python
plt.plot([i[0][:200] for i in res]);
```
